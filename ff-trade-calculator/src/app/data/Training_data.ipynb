```python
import pandas as pd
import numpy as np
```

```python
df08 = pd.read_excel(r"Raw Data 10yrs_v3.xlsx", sheet_name="2008")
df09 = pd.read_excel(r"Raw Data 10yrs_v3.xlsx", sheet_name="2009")
df10 = pd.read_excel(r"Raw Data 10yrs_v3.xlsx", sheet_name="2010")
df11 = pd.read_excel(r"Raw Data 10yrs_v3.xlsx", sheet_name="2011")
df12 = pd.read_excel(r"Raw Data 10yrs_v3.xlsx", sheet_name="2012")
df13 = pd.read_excel(r"Raw Data 10yrs_v3.xlsx", sheet_name="2013")
df14 = pd.read_excel(r"Raw Data 10yrs_v3.xlsx", sheet_name="2014")
df15 = pd.read_excel(r"Raw Data 10yrs_v3.xlsx", sheet_name="2015")
df16 = pd.read_excel(r"Raw Data 10yrs_v3.xlsx", sheet_name="2016")
df17 = pd.read_excel(r"Raw Data 10yrs_v3.xlsx", sheet_name="2017")
df18 = pd.read_excel(r"Raw Data 10yrs_v3.xlsx", sheet_name="2018")

df = [df08,df09,df10,df11,df12,df13,df14,df15,df16,df17,df18]
```

```python
# Rename columns

for i in range(11):
    currDf = df[i]
    currDf.rename(columns={"Unnamed: 0": "Name", "Unnamed: 1": "Team", "Unnamed: 2": "Pos", "Unnamed: 3": "Age", "Games": "G", "Games.1": "GS", "Passing": "P_Cmp", "Passing.1": "P_Att", "Passing.2": "P_Yds", "Passing.3": "P_TD", "Passing.4": "P_Int", "Rushing": "Ru_Att", "Rushing.1": "Ru_Yds", "Rushing.2": "Y/A", "Rushing.3": "R_TD", "Receiving": "Tgt", "Receiving.1": "Rec", "Receiving.2": "Re_Yds", "Receiving.3": "Y/R", "Receiving.4": "Re_TD", "Fantasy": "Total_Pts", "Fantasy.1": "PPR", "Fantasy.2": "PosRank"}, inplace=True)
    df[i] = currDf.drop(currDf.index[0])
    
df[0]
```

```python
# drop the team column and encode the Position column to be numerical

for i in range(11):
    currDf = df[i]
    currDf["Pos"].mask(currDf["Pos"] == "QB", 1, inplace=True)
    currDf["Pos"].mask(currDf["Pos"] == "RB", 2, inplace=True)
    currDf["Pos"].mask(currDf["Pos"] == "WR", 3, inplace=True)
    currDf["Pos"].mask(currDf["Pos"] == "TE", 4, inplace=True)
    currDf.fillna(value=0, axis='columns', inplace=True)
    df[i] = currDf.drop(columns=["Team"])
    
    

df[0]
```

```python
# add column of next season to current season

for i in range(10):
    currDf = df[i]
    nextDf = df[i+1][["Name", "Total_Pts"]]
    nextDf.rename(columns={"Total_Pts": "Next_Season_Pts"}, inplace=True)
    newDf = pd.merge(currDf, nextDf, how="outer", on=["Name"])
#     mask = newDf["Next_Season_Pts"] == 0
#     newDf = newDf[~mask]
    df[i] = newDf
```

```python
# remove players who didn't play a game (this is a result of the join)

for i in range(10):
    currDf = df[i]
    currDf.fillna(value=0, axis='columns', inplace=True)
    for index, row in currDf.iterrows():        
        if row["G"] < 10:
            currDf = currDf.drop(axis=0, index=index)
            
    df[i] = currDf  
    
df[3]
```

```python
# combine the data and export it
del df[10]
combined_df = pd.concat(df, ignore_index=True)

combined_df = combined_df.drop(columns=['PPR', 'PosRank'])
# remove players who scored less than 130 pts (~8pts/game)

for index, row in combined_df.iterrows():        
        if row["Total_Pts"] <= 80 or row["Next_Season_Pts"] <= 80:
            combined_df = combined_df.drop(axis=0, index=index)

file_path = "learning_data.csv"
combined_df.to_csv(file_path, index=False)
```
